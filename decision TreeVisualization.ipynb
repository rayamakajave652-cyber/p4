```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt
import pandas as pd
```

```python
iris = load_iris()

X = iris.data          # Features
y = iris.target        # Labels

feature_names = iris.feature_names
target_names = iris.target_names
```

```python
df = pd.DataFrame(X,columns=feature_names)
df ['species'] = y
df.head()
```

```python
model = DecisionTreeClassifier(max_depth=3)
model.fit(X, y)
```

```python
plt.figure(figsize=(12,8))
plot_tree(model, feature_names=feature_names, class_names=target_names, filled=True)
plt.show()
```

```python
from sklearn.tree import export_text
print(export_text(model,feature_names = feature_names))
```

```python
df ['species'] = df['species'].map({
    0: 'Setosa',
    1: 'versicolor',
    2: 'virginica'
})
df.head()
```

```python
import streamlit as st
import numpy as np

model = pickle.load(open('iris_model.pkl','rb'))

st.title('Iris Prediciton and Visualization')

sepal_length = st.slider("Sepal Length",4.0,8.0,5.0)
sepal_width = st.silder("sepal width",2.0,4.5,3.0)

petal_length = st.slider("Petal Length",1.0,7.0,4.0)
petal_width = st.silder("petal width",0.1,2.5,1.0)

input_data = np.array([[sepal_length,sepal_width,petal_length,petal_width]])
predection = model.predict(input_data)

species = ['setosa','Versicolor','Verginica']

st.header('Prediciton')
st.success(species[prediction[0]])
          
```
